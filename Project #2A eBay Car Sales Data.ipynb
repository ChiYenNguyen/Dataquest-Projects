## Initial Exploration and Cleaning

```python
autos.describe(include='all')
```

'price' and 'odometer' columns have numeric values stored as text. 

```python
print(autos['price'].unique())
print(autos["price"].dtype)
```

```python
autos["price"] = autos["price"].str.replace('$','')
autos["price"] = autos["price"].str.replace(',','')
autos["price"] = autos["price"].astype(int)
```

```python
print(autos["price"].dtype)
print(autos["price"].unique())
```

```python
print(autos["odometer"].unique())
print(autos["odometer"].dtype)
```

```python
autos["odometer"] = autos["odometer"].str.replace('km','')
autos["odometer"] = autos["odometer"].str.replace(',','')
autos["odometer"] = autos["odometer"].astype(int)
print(autos["odometer"].unique())
print(autos["odometer"].dtype)
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
autos["odometer_km"].head()
```

```python
autos = autos.drop(["seller", "offer_type", "nr_of_pictures"], axis=1)
```

## Dealing with Incorrect Registration Year Data

```python
autos["registration_year"].value_counts().sort_index()
```

Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s. Thus, we'll remove those cars that fall outside the 1900 - 2016 interval.

```python
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].describe()
```

```python
# Calculate the distribution of remaining values
autos["registration_year"].value_counts(normalize=True).head(20)
```

It appears that most of the vehicles were first registered in the past 20 years.

## Exploring Price by Brand

```python
print(autos["brand"].unique())
print("\n")
print(autos["brand"].value_counts(normalize=True).head(10))
print("\n")
print(autos["brand"].value_counts(normalize=True).head(10).index)
```

To explore the unique values in the brand column, we choose the first 20 unique brands to aggregate on.

```python
brand_dict = {}

brand_list = autos["brand"].value_counts(normalize=True).head(6).index

for i in brand_list:
    brand_df = autos[autos["brand"] == i]
    mean_price = brand_df["price"].mean()
    brand_dict[i] = int(mean_price)

print(brand_dict)
```

Of the top 6 brands, there is a distinct price gap:

- Audi, BMW and Mercedes Benz are more expensive
- Ford and Opel are less expensive
- Volkswagen is in between - this may explain its popularity

## Storing Aggregate Data in DataFrame

For the top 6 brands, let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price.  While our natural instinct may be to display both aggregated series objects and visually compare them, this has a few limitations:

- it's difficult to compare more than two aggregate series objects if we want to extend to more columns
- we can't compare more than a few rows from each series object
- we can only sort by the index (brand name) of both series objects so we can easily make visual comparisons

Instead, we can combine the data from both series objects into a single dataframe (with a shared index) and display the dataframe directly. 

```python
# Convert dictionary to series object, using the series constructor.
bmp_series = pd.Series(brand_dict)
print(bmp_series)
```

```python
# Create a dataframe from the first series object using the dataframe constructor
df = pd.DataFrame(bmp_series, columns=['mean_price'])
df
```

```python
# Create a dictionary to store mean mileage
brand_mileage = {}

for i in brand_list:
    brand_df1 = autos[autos["brand"] == i]
    mileage = brand_df1["odometer_km"].mean()
    brand_mileage[i] = int(mileage)

print(brand_mileage)
```

```python
# Convert dictionary to series object
brand_mil_series = pd.Series(brand_mileage)
```

```python
# Assign the other series as a new column in this dataframe
df["mean_mileage"] = brand_mil_series
df
```

The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
